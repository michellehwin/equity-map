```python
import pandas as pd
import geohash2
```

```python
csv_name = "Crime_Reports"
df = pd.read_csv(csv_name + ".csv")
df = df.reset_index()

df
```

```python
geohash = []
for index in df.index:
    latitude = df['Latitude'][index]
    longitude = df['Longitude'][index]
    if(latitude and longitude):
        geohash.append(geohash2.encode(
            latitude, longitude))
    else:
        geohash.append(None)
df['geohash'] = geohash
df.to_csv(csv_name+"_Hashed.csv", index=False)
```
